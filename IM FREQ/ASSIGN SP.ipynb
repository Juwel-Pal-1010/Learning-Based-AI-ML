```python
import os
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft, istft, get_window
import scipy.signal as signal
```

```python
audio, sr = librosa.load("SX184.WAV (1).wav")
plt.figure(figsize=(10,6))
plt.plot(audio)
plt.title('Audio signal')
plt.show()
```

```python
audio
```

```python
# Performing FFT & Printing Spectogram From Audio Signal

fft = np.fft.fft(audio)
magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))

plt.figure(figsize=(10, 6))
plt.specgram(audio, Fs=sr, NFFT=256, noverlap=128, cmap='viridis')
plt.title('Spectogram using matplotlib.pyplot.specgram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(label='Intensity (dB)')
plt.show()
```

```python
# Performing STFT & Printing Spectogram From Audio Signal

import librosa.display
X = librosa.stft(audio)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(15, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar(label='Intensity (dB)')
plt.title("Spectogram of STFT")
```

```python
# Performing Inverse Of Both FFT & STFT

# Inverse Of FFT

from scipy.fft import irfft

new_sig = irfft(fft)
plt.plot(new_sig)
plt.show()
```

```python
# Inverse Of STFT

audio_reconstructed = librosa.istft(X)
plt.plot(audio_reconstructed)
plt.show()
```
